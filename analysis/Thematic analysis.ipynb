```python
from collections import Counter
import pandas as pd
```

```python
def basic_stats_categorical(data,name=""):
    print "---------------------"
    print name
    cnt = Counter(data)
#     print cnt
    percentage = np.array(cnt.values())/float(sum(cnt.values()))*100
#     print percentage
    for category,percent in zip(cnt.keys(),percentage):
        print  "{} : %.2f ".format(str(category).ljust(5)) %percent ,"%" 
    sorted_percentage=[]
    for i in sorted(cnt.keys()):
        sorted_percentage.append(cnt[i]/float(sum(cnt.values()))*100)
    return sorted_percentage
```
